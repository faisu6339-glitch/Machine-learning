## Understanding NumPy

**NumPy (Numerical Python)** is a fundamental library for scientific computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of high-level mathematical functions to operate on these arrays efficiently. It is the cornerstone of the Python data science ecosystem, serving as the basis for libraries like Pandas, SciPy, and Scikit-learn.

### Core Features and Concepts:

1.  **`ndarray` (N-dimensional Array) Object:**
    *   The central object in NumPy is the `ndarray`, which is a powerful N-dimensional array object. It's a grid of values, all of the same type, and is indexed by a tuple of non-negative integers.
    *   Unlike Python lists, NumPy arrays are designed to store homogeneous data types, leading to more efficient storage and operations.

2.  **Performance:**
    *   NumPy operations are implemented in C and Fortran, making them significantly faster than equivalent operations on standard Python lists, especially for large datasets.
    *   This efficiency comes from its memory management; `ndarray`s store data in a contiguous block of memory, allowing for optimized vectorized operations.

3.  **Vectorization:**
    *   NumPy encourages writing code using array-based operations (vectorization) instead of explicit loops. This not only makes code more concise and readable but also leverages NumPy's optimized C implementations for dramatic speed improvements.
    *   Example: `arr1 + arr2` performs element-wise addition on two arrays without needing a `for` loop.

4.  **Broadcasting:**
    *   NumPy's broadcasting mechanism allows arithmetic operations between arrays of different shapes. It automatically "stretches" smaller arrays to match the shape of larger arrays, provided they are compatible.
    *   This eliminates the need for explicit tiling or duplication of arrays, saving memory and improving performance.

5.  **Mathematical Functions:**
    *   NumPy provides a vast collection of mathematical functions that operate element-wise on arrays, including:
        *   **Trigonometric:** `sin()`, `cos()`, `tan()`, etc.
        *   **Exponents and Logarithms:** `exp()`, `log()`, `log10()`, etc.
        *   **Rounding:** `floor()`, `ceil()`, `round()`.
        *   **Statistical:** `mean()`, `median()`, `std()`, `sum()`, `max()`, `min()`.
        *   **Linear Algebra:** Dot products, matrix multiplication, determinants, inverses, eigenvalues, etc.

### Why Use NumPy?

*   **Efficiency:** Faster computation on large datasets compared to Python lists.
*   **Memory Efficiency:** Arrays consume less memory than Python lists for storing numerical data.
*   **Convenience:** Rich set of functions for array manipulation and mathematical operations.
*   **Foundation:** Forms the basis for many other scientific computing libraries in Python.

### Basic Array Creation and Manipulation:

```python
import numpy as np

# Creating arrays
arr1 = np.array([1, 2, 3])                 # 1D array
arr2 = np.array([[1, 2], [3, 4]])         # 2D array
arr_zeros = np.zeros((3, 4))              # Array of zeros
arr_ones = np.ones((2, 2))                # Array of ones
arr_range = np.arange(0, 10, 2)           # Array with a range of values
arr_linspace = np.linspace(0, 1, 5)       # Array with evenly spaced values

# Array attributes
print(f"Shape of arr2: {arr2.shape}")        # (rows, columns)
print(f"Dimensions of arr2: {arr2.ndim}")     # Number of dimensions
print(f"Data type of arr2: {arr2.dtype}")    # Data type of elements
print(f"Size of arr2: {arr2.size}")        # Total number of elements

# Array operations
addition = arr1 + arr1
multiplication = arr1 * 2
matrix_mult = np.dot(arr2, arr2) # Matrix multiplication

# Indexing and Slicing
print(arr1[0])          # First element
print(arr2[0, 1])       # Element at row 0, column 1
print(arr1[1:])         # Slice from second element to end
print(arr2[:, 0])       # All rows, first column

# Reshaping
arr_reshaped = np.arange(9).reshape(3, 3)
print(f"Reshaped array:\n{arr_reshaped}")
```

In summary, NumPy is an indispensable library for anyone working with numerical data in Python, providing a robust and efficient framework for array computing.

#3.Program: Student Subject-wise Marks

````python
import numpy as np

# Subject-wise marks of 5 students (5 subjects)
marks = np.array([
    [85, 90, 78, 88, 92],   # Student 1
    [76, 81, 69, 73, 80],   # Student 2
    [95, 91, 89, 94, 90],   # Student 3
    [65, 70, 72, 68, 75],   # Student 4
    [88, 84, 86, 90, 87]    # Student 5
])

# Student names
students = ["Aaliya Fatma", "Shadmani Noor", "Ummehain Noor", "Zainab fatma", "Talat"]

# Subject names
subjects = ["Math", "Science", "English", "Computer", "Physics"]

print("Student Marks Matrix:\n")
print(marks)



# Total marks of each student
total = np.sum(marks, axis=1)

# Average marks of each student
average = np.mean(marks, axis=1)

# Percentage of each student
# Assuming each subject is out of 100
percentage = (total / (marks.shape[1] * 100)) * 100

# -------------------------------
# Display Result
# -------------------------------

print("\nStudent Result\n")
print("-" * 70)

for i in range(len(students)):
    print(f"Student : {students[i]}")
    print("Marks   :", marks[i])
    print(f"Total   : {total[i]}")
    print(f"Average : {average[i]:.2f}")
    print(f"Percent : {percentage[i]:.2f}%")
    print("-" * 70)
````

#4.Program with User Input

````python
import numpy as np

students = int(input("Enter the number of students: "))
subjects = int(input("Enter the number of subjects: "))

# Store student names
student_names = []

for i in range(students):
    name = input(f"Enter name of Student {i+1}: ")
    student_names.append(name)

marks = np.zeros((students, subjects), dtype=int)

# Input marks
for i in range(students):
    print(f"\nEnter marks for {student_names[i]}")

    for j in range(subjects):
        marks[i][j] = int(input(f"Subject {j+1}: "))

# Calculations
total = np.sum(marks, axis=1)
average = np.mean(marks, axis=1)
percentage = (total / (subjects * 100)) * 100

# Display Results
print("\nRESULT")
print("=" * 60)

for i in range(students):
    print(f"Student Name : {student_names[i]}")
    print("Marks        :", marks[i])
    print("Total        :", total[i])
    print("Average      :", round(average[i], 2))
    print("Percentage   :", round(percentage[i], 2), "%")
    print("-" * 60)
````
