#1. np.add()

```python
import numpy as np
arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,9,10])

result=np.add(arr1,arr2)
print(result)
```

#2. np.subtract()

```python
import numpy as np
arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,9,10])

result=np.subtract(arr1,arr2)
print(result)
```

#3. np.multiply()

```python
import numpy as np
arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,9,10])

result=np.multiply(arr1,arr2)
print(result)
```

#4. np.divide()

```python
import numpy as np

a = np.array([20,40,60])
b = np.array([2,4,5])

print(np.divide(a,b))
```

#5. np.mod()

```python
import numpy as np

a = np.array([10,15,20])
b = np.array([3,4,6])

print(np.mod(a,b))
```
