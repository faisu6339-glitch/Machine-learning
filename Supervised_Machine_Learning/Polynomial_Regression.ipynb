## Polynomial Regression Explained

Polynomial Regression is a form of regression analysis in which the relationship between the independent variable \(x\) and the dependent variable \(y\) is modeled as an \(n\)-th degree polynomial. While linear regression assumes a linear relationship between the variables, polynomial regression allows for non-linear relationships, making it more flexible for fitting a wider range of data patterns.

### Mathematical Formulation

The equation for a simple linear regression model is:

\(y = b_0 + b_1x + \epsilon\)

where:
- \(y\) is the dependent variable (output).
- \(x\) is the independent variable (input).
- \(b_0\) is the y-intercept.
- \(b_1\) is the slope of the line.
- \(\epsilon\) is the error term.

In Polynomial Regression, we extend this by adding higher-order polynomial terms:

\(y = b_0 + b_1x + b_2x^2 + \dots + b_nx^n + \epsilon\)

Here:
- \(n\) is the degree of the polynomial.
- \(b_0, b_1, \dots, b_n\) are the coefficients that the model learns.

Despite modeling a non-linear relationship between \(x\) and \(y\), Polynomial Regression is still considered a **linear model** in terms of its coefficients. This is because the equation is linear in the parameters \(b_i\). The techniques used to solve for these coefficients (like Ordinary Least Squares) are the same as those used for linear regression.

### Why use Polynomial Regression?

1.  **Capturing Non-Linearity**: When the relationship between the features and the target variable is clearly non-linear, linear regression will perform poorly. Polynomial regression can effectively model these curved relationships.
2.  **Flexibility**: By increasing the degree of the polynomial, you can fit more complex curves to the data.
3.  **Better Fit**: Often, polynomial regression provides a better fit to the data than simple linear regression, leading to lower residuals and higher accuracy on the training data.

### Challenges and Considerations

1.  **Overfitting**: A significant challenge with polynomial regression is overfitting. As you increase the degree of the polynomial, the model becomes more complex and can fit the training data almost perfectly, including noise. This leads to poor generalization performance on unseen data.
2.  **Choosing the Degree**: Deciding the optimal degree of the polynomial \(n\) is crucial. A low degree might lead to underfitting (the model is too simple), while a high degree might lead to overfitting. Techniques like cross-validation are often used to find the best degree.
3.  **Extrapolation**: Polynomial models can behave erratically when extrapolating (predicting values outside the range of the training data), especially for higher degrees.
4.  **Interpretability**: As the degree of the polynomial increases, the model becomes harder to interpret. The coefficients don't have as straightforward a meaning as in simple linear regression.

### Implementation Steps (General)

1.  **Data Preprocessing**: Prepare your data, handling missing values and scaling if necessary.
2.  **Feature Engineering**: Create new polynomial features from the existing independent variable(s). For example, if you have `x`, you would create `x^2`, `x^3`, etc., up to the desired degree.
3.  **Model Training**: Use a linear regression model (e.g., `LinearRegression` from scikit-learn) with these newly created polynomial features.
4.  **Evaluation**: Evaluate the model's performance using metrics like R-squared, Mean Squared Error (MSE), and visualize the fitted curve against the actual data. Cross-validation is essential to assess generalization.

### Example Use Case

Imagine you are tracking the growth of a plant over time. Initially, the growth might be slow, then it accelerates, and finally, it slows down again as the plant reaches maturity. A linear model would struggle to capture this S-shaped curve, but a polynomial regression model (e.g., degree 3 or 4) could fit this pattern much more accurately.

#Program 1: Student Study Hours vs Marks Prediction

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Dataset
X = np.array([[1],[2],[3],[4],[5],[6],[7],[8]])
y = np.array([35,40,50,65,80,90,95,98])
```

```python
poly=PolynomialFeatures(degree=2)
X_poly=poly.fit_transform(X)
```

```python
model=LinearRegression()
model.fit(X_poly,y)
```

```python
# Prediction
hours = float(input("Enter Study Hours: "))
prediction = model.predict(poly.transform([[hours]]))

print("Predicted Marks:", prediction[0])
```

```python
# Graph
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X_poly), color='red')
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Polynomial Regression")
plt.show()
```

#Program 2: Salary Prediction Based on Experience

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

X = np.array([[1],[2],[3],[4],[5],[6],[7]])
y = np.array([30000,35000,42000,52000,68000,85000,105000])
```

```python
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly,y)
```

```python
exp = float(input("Enter Experience: "))

salary = model.predict(poly.transform([[exp]]))

print("Predicted Salary =", salary[0])
```

#Program 4: Car Price Prediction

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

X = np.array([[1],[2],[3],[4],[5],[6],[7],[8]])
y = np.array([900000,850000,790000,700000,600000,480000,360000,250000])

poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly,y)

age = float(input("Enter Car Age: "))

price = model.predict(poly.transform([[age]])

)

print("Predicted Car Price =", int(price[0]))
```
