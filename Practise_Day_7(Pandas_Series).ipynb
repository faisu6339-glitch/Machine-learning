```python
import pandas as pd
import numpy as np
```

#Creating Series

#1. Create a Series from a List

```python
data=[10,20,30,40,50]
s=pd.Series(data)
print(s)
```

#2. Create Series from Tuple

```python
import pandas as pd

data = (100,200,300,400)

s = pd.Series(data)

print(s)
```

#3. Create Series from NumPy Array

```python
import pandas as pd
import numpy as np

arr = np.array([5,10,15,20])

s = pd.Series(arr)

print(s)
```

#4. Create Series from Dictionary

```python
import pandas as pd

student = {
    "Math":90,
    "Science":85,
    "English":95
}

s = pd.Series(student)

print(s)
```

#5. Create Series using Scalar Value

```python
import pandas as pd

s = pd.Series(100,index=['A','B','C','D'])

print(s)
```

#6. Create Empty Series

```python
import pandas as pd

s = pd.Series(dtype='float64')

print(s)
```

#INDEX

#7.Display Index

```python
import pandas as pd

s = pd.Series([10,20,30,40])

print(s.index)
```

#8. Create Series with Custom Index

```python
import pandas as pd

s = pd.Series([80,90,85],index=['Math','Science','English'])

print(s)
```

#ACCESSING ELEMENTS

#9. Access using Custom Index

```python
import pandas as pd

s = pd.Series([80,90,85],index=['Math','Science','English'])

print(s['Science'])
```

#10. Access by Position

```python
import pandas as pd

s = pd.Series([10,20,30,40])

print(s[2])
```

```python
import pandas as pd

s = pd.Series([10,20,30,40])

print(s[2])
```

```python
import pandas as pd

s = pd.Series([5,10,15,20])

print(s.iloc[2])
```

```python
import pandas as pd

s = pd.Series([100,200,300],index=['A','B','C'])

print(s.loc['B'])
```

#SLICING

#14. Slice using Position

```python
import pandas as pd

s = pd.Series([10,20,30,40,50])

print(s[1:4])
```

#15. Slice using iloc

```python
import pandas as pd

s = pd.Series([10,20,30,40,50])

print(s.iloc[1:3])
```

#16. Slice using Labels

```python
import pandas as pd

s = pd.Series([10,20,30,40],index=['A','B','C','D'])

print(s.loc['B':'D'])
```

#UPDATING VALUES

#17. Update One Value

```python
import pandas as pd

s = pd.Series([10,20,30])

s[1]=100

print(s)
```

#18. Update Multiple Values

```python
import pandas as pd

s = pd.Series([10,20,30,40])

s[[0,2]]=999,88

print(s)
```

#19. Update using Custom Index

```python
import pandas as pd

s = pd.Series([90,80,70],index=['Math','Science','English'])

s['Science']=100

print(s)
```
