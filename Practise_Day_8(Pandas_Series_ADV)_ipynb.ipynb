#1. Student Management DataFrame

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Roll":[101,102,103,104,105],
    "Name":["Aaliya","Zainab","Shadmani","Ummehani","Talat"],
    "Math":[90,80,70,85,60],
    "Science":[88,75,65,95,70],
    "English":[85,79,72,80,68]
})
```

```python
df
```

```python
df["Total"] = df[["Math","Science","English"]].sum(axis=1)
df["Average"] = df[["Math","Science","English"]].mean(axis=1)
```

```python
df["Grade"] = df["Average"].apply(
    lambda x: "A" if x>=80 else
              "B" if x>=70 else
              "C" if x>=60 else
              "D"
)
```

```python
print(df[["Name", "Total"]])
```

```python
for index, row in df.iterrows():
    print(f"{row['Name']} : {row['Total']}")
```

```python
print(df[["Name", "Total", "Grade"]])
```

```python
for index, row in df.iterrows():
    print(f"Student Name: {row['Name']}, Total Marks: {row['Total']}")
```

```python
print(df[["Name","Math"]])
```

#Update a student's name

```python
df.loc[df["Name"]=="Aaliya","Name"]="Aaliya Fatma"
print(df)
```

#Update Math marks

```python
df.loc[df["Name"]=="Aaliya Fatma","Math"] = 90

print(df)
```

#Update multiple columns

```python
df.loc[df["Name"]=="Zainab",["Math","Science"]] = [90,95]

print(df)
```

#Update using Roll Number

```python
df.loc[df["Roll"]==105,"English"] = 90

print(df)
```

#Increase everyone's Math marks by 5

```python
df["Math"] = df["Math"] + 5

print(df)
```

#Add New Data

```python
df["Computer"] = [90,88,75,80,95]

print(df)
```

#Add another column

```python
df["City"] = ["Patna","Delhi","Mumbai","Pune","Kolkata"]

print(df)
```

```python
df = df[
    [
        "Roll",
        "Name",
        "Math",
        "Science",
        "English",
        "Computer",
        "Total",
        "Average",
        "City",
        "Grade"
    ]
]

print(df)
```

#Add one student

```python
new_student = {
    "Roll":106,
    "Name":"Md Faisal Sheikh",
    "Math":89,
    "Science":90,
    "English":85,
    "Total":264,
    "Average":88,
    "Grade":"B"
}

df.loc[len(df)] = new_student

print(df)
```

#Delete multiple columns

```python
df.drop(["Math","Science"],axis=1,inplace=True)

print(df)
```

```python
df.insert(3, "Math", [90, 88, 75, 80, 95, 82])

print(df)
```

#Rename Columns

```python
df.rename(columns={"Math":"Mathematics"},inplace=True)

print(df)
```

#Insert Column at Specific Position

```python
df.insert(2,"Gender",["F","F","F","F","F","M"])
print(df)
```

#Select Students

```python
print(df[df["Total"]>250])
```
