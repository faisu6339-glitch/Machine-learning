#Workking with Dates

converting to datetime datatype

```python
date['order_date']=pd.to_datetime(date['order_date'])
```

```python
date.info()
```

Extract year

```python
date['date_year']=date['order_date'].dt.year
```

```python
date.sample(5)
```

Extract Months

```python
date['date_month_no']=date['order_date'].dt.month
```

```python
date.head()
```

```python
date['date_month_name']=date['order_date'].dt.month_name()
```

```python
date.head(5)
```

Extract Days

```python
date['date_day']=date['order_date'].dt.day
```

```python
date.head(5)
```

Day of week

```python
date['date_dow']=date['order_date'].dt.dayofweek
```

```python
date.head()
```

Day of week -name

```python
date['date_dow_name']=date['order_date'].dt.day_name()
date.drop(columns=['product_id','city_id','order_id','price','quantity','total_amount']).head()
```

Is weekend

```python
date['date_is_weekend']=np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['product_id','city_id','order_id','price','quantity','total_amount']).head()
```

Extract week of the year

```python
date['date_week']=date['order_date'].dt.isocalendar().week
date.drop(columns=['product_id','city_id','order_id','price','quantity','total_amount']).head()
```

Extract Quarter

```python
date['date_quarter']=date['order_date'].dt.quarter
date.drop(columns=['product_id','city_id','order_id','price','quantity','total_amount']).head()
```

Extract Semester

```python
date['date_semester'] = np.where(date['date_month_no'].isin([1, 2, 3, 4, 5, 6]), 1, 2)
date.drop(columns=['product_id','city_id','order_id','price','quantity','total_amount']).head()
```

```python
# For demonstration, let's create a hypothetical 'delivery_date' column
# by adding a random number of days (1 to 10) to the 'order_date'.
# In a real scenario, you would use your actual second date column.
import datetime

date['delivery_date'] = date['order_date'] + pd.to_timedelta(np.random.randint(1, 11, size=len(date)), unit='D')

# Calculate the time elapsed in days
date['time_elapsed_days'] = (date['delivery_date'] - date['order_date']).dt.days

# Display the relevant columns to show the result
display(date[['order_date', 'delivery_date', 'time_elapsed_days']].head())
```

```python
import datetime
today = pd.to_datetime(datetime.date.today())
(today - date['order_date']).dt.days
```

Months passed

```python
# Calculate total months for today
total_months_today = today.year * 12 + today.month

# Calculate total months for each order_date
total_months_order_date = date['order_date'].dt.year * 12 + date['order_date'].dt.month

# Calculate the difference in months
months_passed = total_months_today - total_months_order_date

# Rounding to 0 decimal places as in the original request
np.round(months_passed, 0)
```
