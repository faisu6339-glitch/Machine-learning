#1. concatenate()

```python
import numpy as np
arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,9,10])

result=np.concatenate((arr1,arr2))
print(result)
```

2D Arrays (Row Wise)

```python
a=np.array([[1,2],
           [3,4]])
b=np.array([[5,6],
           [7,8]])

c=np.concatenate((a,b),axis=0)
print(c)
```

Column Wise

```python
a=np.array([[1,2],
           [3,4]])
b=np.array([[5,6],
           [7,8]])

c=np.concatenate((a,b),axis=1)
print(c)
```

#2. stack()

```python
a=np.array([1,2,3])
b=np.array([4,5,6])

c=np.stack((a,b))
print(c)
```

```python
import numpy as np

a = np.array([1,2,3])

b = np.array([4,5,6])

result = np.stack((a,b), axis=1)

print(result)
```

#3. vstack()

```python
a=np.array([1,2,3])
b=np.array([4,5,6])

c=np.vstack((a,b))
print(c)
```

```python
import numpy as np

a = np.array([[1,2],
              [3,4]])

b = np.array([[5,6]])

result = np.vstack((a,b))

print(result)
```

#4. hstack()

```python
a=np.array([1,2,3])
b=np.array([4,5,6])

c=np.hstack((a,b))
print(c)
```

```python
import numpy as np

a = np.array([[1,2],
              [3,4]])

b = np.array([[5],
              [6]])

result = np.hstack((a,b))

print(result)
```

#Multiple Columns

```python
import numpy as np

a = np.array([1,2,3])

b = np.array([4,5,6])

c = np.array([7,8,9])

result = np.column_stack((a,b,c))

print(result)
```

#7. row_stack()

```python
import numpy as np

a = np.array([1,2,3])

b = np.array([4,5,6])

result = np.row_stack((a,b))

print(result)
```

#8. block()

```python
import numpy as np

A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])

result = np.block([[A,B]])

print(result)
```

#Practical Program 1: Join Two Student Marks

```python
import numpy as np

student1 = np.array([80,85,90])

student2 = np.array([75,88,92])

all_marks = np.concatenate((student1,student2))

print(all_marks)
```

#Practical Program 2: Combine Two Classes

```python
import numpy as np

classA = np.array([[80,90],
                   [75,88]])

classB = np.array([[92,95],
                   [78,84]])

all_students = np.vstack((classA,classB))

print(all_students)
```
